```python
#Tratamiento de Datos y Primero Gráficos

import pandas as pd
datos = pd.read_csv('inmuebles_bogota.csv', header=(0))
datos
```

```python
datos.info()
```

# Nueva sección

```python
columna = {
    'Baños': 'Banos',
    'Área': 'Area'
}

datos = datos.rename(columns=columna)
datos.sample(5)
```

```python
#Como convertir de str a int los datos de las tabla

type(datos.Valor[0])
```

```python
datos.Valor[0]+datos.Valor[1]+datos.Valor[:5]
```

```python
datos.Valor[0].split()
```

```python
valor = datos.Valor.str.split(expand=True)
datos['Moneda'] = valor[0]
datos['Precio'] = valor[1]
datos.sample(3)
```

```python
datos.info()
```

```python
datos['Precio'].sum()
```

```python
datos['Precio'] = datos['Precio'].str.replace('.','', regex=True)
```

```python
datos[['Precio', 'Barrio']]
```

```python
datos.info()
```

```python
datos['Precio_Millon'] = datos.Precio.astype('float')/1000000
datos.info()
```

```python
datos.describe()
```

```python
pd.set_option('display.precision', 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
datos.describe()
```

```python
datos.loc[datos.Habitaciones == 110]
```

```python
datos.loc[datos.Area == 2]
```

```python
datos['Precio_Millon'].plot.hist(bins=10)
```

```python
#graficas
#seaborn es una libreria de graficas

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
grafica = sns.histplot(data=datos, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de valores de los inmuebles en bogota')
plt.xlim((50,1000))
plt.ylim((0,1000))
plt.savefig('ruta de la imagen en png', format='png')
plt.show()
```
